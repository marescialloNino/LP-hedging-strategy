```python
import pandas as pd

def event_list_to_df(list_of_dict, event_type):
    """Convert event list to DataFrame with event_type column."""
    if not list_of_dict:
        df = pd.DataFrame(columns=['tx_id', 'position_address', 'pair_address', 'token_x_amount', 
                                   'token_y_amount', 'token_x_usd_amount', 'token_y_usd_amount', 
                                   'onchain_timestamp', 'datetime', 'event_type'])
    else:
        df = pd.DataFrame(list_of_dict)
        df = df.sort_values(by='onchain_timestamp')
        df['total_usd_value'] = df['token_x_usd_amount'] + df['token_y_usd_amount']
        df['datetime'] = pd.to_datetime(df['onchain_timestamp'], unit='s')
    df['event_type'] = event_type
    return df

# Create position_history_df and add adjusted token amounts
deposits_df = event_list_to_df(deposits, 'Deposit')
withdraws_df = event_list_to_df(withdraws, 'Withdrawal')
claimfees_df = event_list_to_df(claim_fees, 'Claim Fees')
position_history_df = pd.concat([deposits_df, withdraws_df, claimfees_df], ignore_index=True)
position_history_df = position_history_df.sort_values(by='onchain_timestamp')

def add_adjusted_amounts(df, token_x_decimals, token_y_decimals):
    """Add decimal-adjusted token amounts to the DataFrame."""
    df['token_x_amount_adjusted'] = df['token_x_amount'] / (10 ** token_x_decimals)
    df['token_y_amount_adjusted'] = df['token_y_amount'] / (10 ** token_y_decimals)
    return df

def calculate_pnl_usd_from_history(position_history_df, current_position_value_usd, 
                                  token_x_decimals, token_y_decimals):
    """
    Calculate PnL in USD with adjusted token amounts shown in DataFrame.
    """
    # Add adjusted amounts
    position_history_df = add_adjusted_amounts(position_history_df, token_x_decimals, token_y_decimals)
    
    # Initialize columns for tracking
    position_history_df['cumulative_claimed_x'] = (
        position_history_df[position_history_df['event_type'] == 'Claim Fees']['token_x_amount']
        .cumsum().reindex(position_history_df.index, fill_value=0)
    )
    position_history_df['cumulative_claimed_y'] = (
        position_history_df[position_history_df['event_type'] == 'Claim Fees']['token_y_amount']
        .cumsum().reindex(position_history_df.index, fill_value=0)
    )
    
    # Track available tokens and other metrics
    position_history_df['available_claimed_x'] = 0
    position_history_df['available_claimed_y'] = 0
    position_history_df['reinvested_x'] = 0
    position_history_df['reinvested_y'] = 0
    position_history_df['cost_basis'] = 0
    
    available_x = 0
    available_y = 0
    
    for idx, row in position_history_df.iterrows():
        if row['event_type'] == 'Claim Fees':
            available_x = position_history_df.at[idx, 'cumulative_claimed_x']
            available_y = position_history_df.at[idx, 'cumulative_claimed_y']
        elif row['event_type'] == 'Deposit':
            deposit_x_adjusted = row['token_x_amount_adjusted']
            deposit_y_adjusted = row['token_y_amount_adjusted']
            deposit_x = row['token_x_amount']
            deposit_y = row['token_y_amount']
            token_x_usd = row['token_x_usd_amount']
            token_y_usd = row['token_y_usd_amount']
            
            # Adjust available amounts
            available_x_adjusted = available_x / (10 ** token_x_decimals)
            available_y_adjusted = available_y / (10 ** token_y_decimals)
            
            # Calculate reinvested ratios using adjusted amounts
            reinvested_ratio_x = available_x_adjusted / deposit_x_adjusted if deposit_x_adjusted > 0 else 0
            reinvested_ratio_x = min(reinvested_ratio_x, 1)
            reinvested_ratio_y = available_y_adjusted / deposit_y_adjusted if deposit_y_adjusted > 0 else 0
            reinvested_ratio_y = min(reinvested_ratio_y, 1)
            
            # Split USD values
            reinvested_x_usd = token_x_usd * reinvested_ratio_x
            capital_x_usd = token_x_usd * (1 - reinvested_ratio_x)
            reinvested_y_usd = token_y_usd * reinvested_ratio_y
            capital_y_usd = token_y_usd * (1 - reinvested_ratio_y)
            
            # Total cost basis
            cost_basis = capital_x_usd + capital_y_usd
            
            # Reinvested token quantities (in raw units)
            reinvested_x = deposit_x * reinvested_ratio_x
            reinvested_y = deposit_y * reinvested_ratio_y
            
            # Update DataFrame
            position_history_df.at[idx, 'available_claimed_x'] = available_x
            position_history_df.at[idx, 'available_claimed_y'] = available_y
            position_history_df.at[idx, 'reinvested_x'] = reinvested_x
            position_history_df.at[idx, 'reinvested_y'] = reinvested_y
            position_history_df.at[idx, 'cost_basis'] = cost_basis
            
            # Update available tokens
            available_x = max(0, available_x - reinvested_x)
            available_y = max(0, available_y - reinvested_y)
        else:  # Withdrawal or other
            position_history_df.at[idx, 'available_claimed_x'] = available_x
            position_history_df.at[idx, 'available_claimed_y'] = available_y
    
    # Total capital deposits
    capital_deposits_usd = position_history_df['cost_basis'].sum()
    
    # Total deposits, fees, and withdrawals
    total_deposit_usd = position_history_df[position_history_df['event_type'] == 'Deposit'][['token_x_usd_amount', 'token_y_usd_amount']].sum().sum()
    total_fee_reward_usd = position_history_df[position_history_df['event_type'] == 'Claim Fees'][['token_x_usd_amount', 'token_y_usd_amount']].sum().sum()
    total_withdrawal_usd = position_history_df[position_history_df['event_type'] == 'Withdrawal'][['token_x_usd_amount', 'token_y_usd_amount']].sum().sum()
    
    # Reinvested fees
    reinvested_fees_usd = total_deposit_usd - capital_deposits_usd
    
    # Total inflows
    total_inflows_usd = capital_deposits_usd + total_fee_reward_usd + reinvested_fees_usd
    
    # PnL calculations
    withdrawal_ratio = total_withdrawal_usd / total_inflows_usd if total_inflows_usd > 0 else 0
    withdrawal_ratio = min(withdrawal_ratio, 1)
    withdrawn_capital_usd = capital_deposits_usd * withdrawal_ratio
    realized_pnl_usd = total_withdrawal_usd - withdrawn_capital_usd
    remaining_capital_usd = capital_deposits_usd - withdrawn_capital_usd
    unrealized_pnl_usd = current_position_value_usd - remaining_capital_usd
    net_pnl_usd = (current_position_value_usd + total_withdrawal_usd) - capital_deposits_usd
    
    return {
        'realized_pnl_usd': realized_pnl_usd,
        'unrealized_pnl_usd': unrealized_pnl_usd,
        'net_pnl_usd': net_pnl_usd,
        'capital_deposits_usd': capital_deposits_usd,
        'reinvested_fees_usd': reinvested_fees_usd,
        'total_fee_reward_usd': total_fee_reward_usd
    }

def calculate_pnl_token_y_from_history(position_history_df, current_position_value_usd, 
                                      current_token_y_price_usd, token_x_decimals, token_y_decimals):
    """
    Calculate PnL in Token Y with adjusted token amounts.
    """
    def convert_to_token_y(row, current_price):
        if row['token_y_amount'] == 0:
            return (row['token_x_usd_amount'] + row['token_y_usd_amount']) / current_price
        actual_token_y_quantity = row['token_y_amount'] / (10 ** token_y_decimals)
        token_y_price_at_event = row['token_y_usd_amount'] / actual_token_y_quantity
        return (row['token_x_usd_amount'] + row['token_y_usd_amount']) / token_y_price_at_event
    
    # Add adjusted amounts and Token Y value
    position_history_df = add_adjusted_amounts(position_history_df, token_x_decimals, token_y_decimals)
    position_history_df['token_y_value'] = position_history_df.apply(
        lambda row: convert_to_token_y(row, current_token_y_price_usd), axis=1)
    
    # Initialize columns for tracking
    position_history_df['cumulative_claimed_x'] = (
        position_history_df[position_history_df['event_type'] == 'Claim Fees']['token_x_amount']
        .cumsum().reindex(position_history_df.index, fill_value=0)
    )
    position_history_df['cumulative_claimed_y'] = (
        position_history_df[position_history_df['event_type'] == 'Claim Fees']['token_y_amount']
        .cumsum().reindex(position_history_df.index, fill_value=0)
    )
    
    # Track available tokens and cost basis
    position_history_df['available_claimed_x'] = 0
    position_history_df['available_claimed_y'] = 0
    position_history_df['reinvested_x'] = 0
    position_history_df['reinvested_y'] = 0
    position_history_df['cost_basis_token_y'] = 0
    
    available_x = 0
    available_y = 0
    
    for idx, row in position_history_df.iterrows():
        if row['event_type'] == 'Claim Fees':
            available_x = position_history_df.at[idx, 'cumulative_claimed_x']
            available_y = position_history_df.at[idx, 'cumulative_claimed_y']
        elif row['event_type'] == 'Deposit':
            deposit_x_adjusted = row['token_x_amount_adjusted']
            deposit_y_adjusted = row['token_y_amount_adjusted']
            deposit_x = row['token_x_amount']
            deposit_y = row['token_y_amount']
            token_x_usd = row['token_x_usd_amount']
            token_y_usd = row['token_y_usd_amount']
            deposit_token_y = row['token_y_value']
            
            # Adjust available amounts
            available_x_adjusted = available_x / (10 ** token_x_decimals)
            available_y_adjusted = available_y / (10 ** token_y_decimals)
            
            # Calculate reinvested ratios
            reinvested_ratio_x = available_x_adjusted / deposit_x_adjusted if deposit_x_adjusted > 0 else 0
            reinvested_ratio_x = min(reinvested_ratio_x, 1)
            reinvested_ratio_y = available_y_adjusted / deposit_y_adjusted if deposit_y_adjusted > 0 else 0
            reinvested_ratio_y = min(reinvested_ratio_y, 1)
            
            # Split USD values and convert to Token Y
            reinvested_x_usd = token_x_usd * reinvested_ratio_x
            capital_x_usd = token_x_usd * (1 - reinvested_ratio_x)
            reinvested_y_usd = token_y_usd * reinvested_ratio_y
            capital_y_usd = token_y_usd * (1 - reinvested_ratio_y)
            cost_basis_token_y = (capital_x_usd + capital_y_usd) / current_token_y_price_usd
            
            # Reinvested token quantities (raw units)
            reinvested_x = deposit_x * reinvested_ratio_x
            reinvested_y = deposit_y * reinvested_ratio_y
            
            # Update DataFrame
            position_history_df.at[idx, 'available_claimed_x'] = available_x
            position_history_df.at[idx, 'available_claimed_y'] = available_y
            position_history_df.at[idx, 'reinvested_x'] = reinvested_x
            position_history_df.at[idx, 'reinvested_y'] = reinvested_y
            position_history_df.at[idx, 'cost_basis_token_y'] = cost_basis_token_y
            
            # Update available tokens
            available_x = max(0, available_x - reinvested_x)
            available_y = max(0, available_y - reinvested_y)
        else:
            position_history_df.at[idx, 'available_claimed_x'] = available_x
            position_history_df.at[idx, 'available_claimed_y'] = available_y
    
    # Total capital deposits
    capital_deposits_token_y = position_history_df['cost_basis_token_y'].sum()
    
    # Total deposits, fees, and withdrawals
    total_deposit_token_y = position_history_df[position_history_df['event_type'] == 'Deposit']['token_y_value'].sum()
    total_fee_reward_token_y = position_history_df[position_history_df['event_type'] == 'Claim Fees']['token_y_value'].sum()
    total_withdrawal_token_y = position_history_df[position_history_df['event_type'] == 'Withdrawal']['token_y_value'].sum()
    
    # Reinvested fees
    reinvested_fees_token_y = total_deposit_token_y - capital_deposits_token_y
    
    # Total inflows
    total_inflows_token_y = capital_deposits_token_y + total_fee_reward_token_y + reinvested_fees_token_y
    
    # PnL calculations
    withdrawal_ratio = total_withdrawal_token_y / total_inflows_token_y if total_inflows_token_y > 0 else 0
    withdrawal_ratio = min(withdrawal_ratio, 1)
    withdrawn_capital_token_y = capital_deposits_token_y * withdrawal_ratio
    realized_pnl_token_y = total_withdrawal_token_y - withdrawn_capital_token_y
    current_position_value_token_y = current_position_value_usd / current_token_y_price_usd
    remaining_capital_token_y = capital_deposits_token_y - withdrawn_capital_token_y
    unrealized_pnl_token_y = current_position_value_token_y - remaining_capital_token_y
    net_pnl_token_y = (current_position_value_token_y + total_withdrawal_token_y) - capital_deposits_token_y
    
    return {
        'realized_pnl_token_y': realized_pnl_token_y,
        'unrealized_pnl_token_y': unrealized_pnl_token_y,
        'net_pnl_token_y': net_pnl_token_y,
        'capital_deposits_token_y': capital_deposits_token_y,
        'reinvested_fees_token_y': reinvested_fees_token_y
    }

# Example usage
current_position_value_usd = 372
current_token_y_price_usd = 0.009
token_x_decimals = 6  # Adjust based on your Token X
token_y_decimals = 6  # From your earlier example

usd_pnl = calculate_pnl_usd_from_history(position_history_df, current_position_value_usd, 
                                        token_x_decimals, token_y_decimals)
token_y_pnl = calculate_pnl_token_y_from_history(position_history_df, current_position_value_usd, 
                                                current_token_y_price_usd, token_x_decimals, token_y_decimals)

print("USD PnL:")
print(usd_pnl)
print("\nToken Y PnL:")
print(token_y_pnl)

# Display history with adjusted amounts
print("\nPosition History:")
print(position_history_df[['datetime', 'event_type', 'token_x_amount_adjusted', 'token_y_amount_adjusted', 
                          'token_x_usd_amount', 'token_y_usd_amount', 'total_usd_value', 
                          'available_claimed_x', 'available_claimed_y', 
                          'reinvested_x', 'reinvested_y', 'cost_basis']])
```
